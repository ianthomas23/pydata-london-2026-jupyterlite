# Example using matplotlib and ipympl for interactive plotting

```python
%matplotlib ipympl
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x, y = np.meshgrid(np.linspace(0, 2*np.pi, 20), np.linspace(0, 2*np.pi, 20))
z = np.sin(x - y/4)*np.sin(y)
cs = ax.contourf(x, y, z)
fig.colorbar(cs, ax=ax);
```
