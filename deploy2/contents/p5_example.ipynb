# 🎨 Interactive Circles¶

This is a simple but fun interactive p5.js sketch! Move your mouse over the canvas to paint with colorful, randomly-sized circles.

Features
Mouse tracking: Circles appear wherever your cursor goes
Random colors: Each circle gets a unique RGB color
Transparency: The semi-transparent fill creates beautiful overlapping effects
Variable size: Circle sizes range from 20 to 50 pixels
This example shows how easy it is to create interactive visuals with p5.js!

```p5js
function setup() {
  createCanvas(innerWidth, innerHeight);
  background(30);
}

function draw() {
  // Colorful circles following the mouse
  noStroke();
  fill(random(255), random(255), random(255), 100);
  circle(mouseX, mouseY, random(20, 50));
}
```
